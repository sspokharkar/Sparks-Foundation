```python
#importing required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/Dataset/globalterrorismdb_0718dist.csv", encoding='ISO-8859-1')       #Importing dataset
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df1 = df.copy()
```

```python
df1.head()
```

```python
df1.dropna(axis=1,how='all')
```

```python
df1.shape
```

```python
df2 = df1[['eventid','iyear','imonth','iday','country_txt','region_txt','provstate','city','success',
          'attacktype1_txt','targtype1_txt','target1','nkill','nwound','summary','gname','weaptype1_txt','motive']]
```

```python
df2
```

```python
df2.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State','city':'City','success':'Success','region_txt':'Region',
                    'attacktype1_txt':'Attacktype','target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                    'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
df2.shape
```

```python
df2['casualities']=df2['Killed']+df2['Wounded']
```

```python
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2.describe()
```

#**Number Of Terrorist Activities Each Year**

```python
plt.subplots(figsize=(15,6))
sns.countplot('Year',data=df2,palette='RdYlGn_r',edgecolor=sns.color_palette("YlOrBr",10))
plt.xticks(rotation=90)
plt.title('Number of terrorist activites each year',fontsize=20)
plt.show
```

From above plot we can conclude that the maximum attacks took place in year 2014

#**Number of Terrorist Activities by Region in Each Year**

```python
#Terrorist Activities by Region
pd.crosstab(df2.Year,df2.Region).plot(kind='area',figsize=(18,10))
plt.title('Terrorist Activities by Region in Each Year',fontsize=20)
plt.ylabel('Number of Attacks')
plt.show()
```

#**Attacking Methods by Terrorists**

```python
plt.subplots(figsize=(15,6))
sns.countplot('Attacktype',data=df2,palette='rocket',order=df2['Attacktype'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Attacking methods by terrorists',fontsize=20)
plt.show()
```

#**States with Terrorist Activities**

```python
from wordcloud import WordCloud
from scipy import signal
cities = df2.State.dropna(False)
plt.subplots(figsize=(20,20))
wordcloud=WordCloud(background_color='Black',width=512,height=382).generate(' '.join(cities))
plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```

From above chart we can conclude that the maximum attacks were carried out by bombing/Explosion

#**Countries affected by Terror Activities**

```python
#Top Countries affected by Terror Activities
plt.subplots(figsize=(18,6))
sns.barplot(df2['Country'].value_counts()[:15].index,df2['Country'].value_counts()[:15].values)
plt.title('Top Countries Affected',fontsize=20)
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

#**Favourite Targets of Terrorists**

```python
plt.subplots(figsize=(15,6))
sns.countplot('Target_type',data=df2,palette='mako',order=df2['Target_type'].value_counts().index)
plt.xticks(rotation=90)
plt.title("Favorite targets",fontsize=20)
plt.show()
```

#**Year wise increase in terror activities in regions**

```python
terror_region=pd.crosstab(df2.Year,df2.Region)
terror_region.plot(color=sns.color_palette('Paired',12))
fig=plt.gcf()
fig.set_size_inches(18,6)
plt.title('Year wise increase in terror activities in regions',fontsize=20)
plt.show
```

#**Attacks in Countries using Globe**

```python
ter = df2.groupby(['Country'],as_index=False).count()
import plotly.express as px
fig = px.choropleth(ter, locations='Country',locationmode='country names',
                    color='Year',hover_name='Country',projection='orthographic',
                    title='Total number of attacks (1970-2017)', labels={'Year':'Attacks'})
fig.show()
```

#**Number of attacks took place and number of people killed**

```python
count_terror = df2['Country'].value_counts()[:15].to_frame()
count_terror.columns=['Attacks']
coun_kill = df2.groupby('Country')['Killed'].sum().to_frame()
count_terror.merge(coun_kill, left_index=True,right_index=True,how='left').plot.bar(width=0.9)
fig=plt.gcf()
fig.set_size_inches(18,6)
plt.title('Number of attacks took place and number of people killed',fontsize=15)
plt.show
```

#**Number of attacks took place and number of peoples killed and wounded**

```python
count_terror = df2['Country'].value_counts()[:15].to_frame()
count_terror.columns=['Attacks']
coun_kill = df2.groupby('Country')['casualities'].sum().to_frame()
count_terror.merge(coun_kill, left_index=True,right_index=True,how='left').plot.bar(width=0.9)
fig=plt.gcf()
fig.set_size_inches(18,6)
plt.title('Number of attacks took place and number of people killed and wounded',fontsize=15)
plt.show
```

#**Most Active Organiztions**

```python
#Most active organizations
sns.barplot(df2['Group'].value_counts()[1:20].values,df2['Group'].value_counts()[1:20].index,palette=('copper'))
plt.xticks(rotation=90)
fig=plt.gcf()
fig.set_size_inches(10,8)
plt.title('Most Active Organizations',fontsize=15)
plt.show()
```
